```python
import numpy as np
import pandas as pd
import matplotlib as mtl
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns

data=pd.read_csv('flipkart.csv')
X=data.drop(columns=['product_name','product_url','uniq_id','crawl_timestamp','product_category_tree','pid','image','is_FK_Advantage_product','overall_rating',
                            'brand','product_specifications','product_rating','description'])
Y=data['product_category_tree']

dataset=dataset=pd.concat([X,Y],axis=1)
dataset
```

```python

dataset.isnull().values.any()
dataset['retail_price'].fillna((dataset['retail_price'].mean()),inplace=True)
dataset['discounted_price'].fillna((dataset['discounted_price'].mean()),inplace=True)
```

```python
dataset.duplicated().shape
```

```python
new_x=dataset.iloc[:,0:2]
new_x
```

```python
new_y=dataset.iloc[:,2:]
new_y
```

```python

lb=LabelEncoder()
dataset['product_category_tree']=lb.fit_transform(dataset['product_category_tree'])
dataset
```

```python

modal= tree.DecisionTreeClassifier()

modal.fit(new_x,new_y)
predection=modal.predict([ [999.0,271] ])
predection
```

```python
#accuracy measurment
x_train,x_test,y_train,y_test=(train_test_split(new_x,new_y,test_size=0.1000))
modal.fit(x_train,y_train)
predection=modal.predict(x_test)
score=accuracy_score(y_test,predection)
print('score=',score)
```

```python
#visualization
from matplotlib import pyplot as plt
plt.plot(dataset.retail_price,dataset.discounted_price,dataset.product_category_tree,) 
plt.xlabel('new_x')
plt.ylabel('new_y')
plt.show()
```
